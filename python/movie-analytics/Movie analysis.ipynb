```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 20)
pd.set_option('display.max_rows', 25)
```

```python
dir_name = 'movie-data/'
movies_table=pd.read_table(dir_name + 'movies.dat',sep='::',header=None,names=['movie_id','title','genres'],engine='python').set_index('movie_id')
ratings_table=pd.read_table(dir_name + 'ratings.dat',sep='::',header=None,names=['user_id','movie_id','rating','timestamp'],engine='python')
```

```python
movies_table.info()
```

```python
ratings_table
```

```python
ratings_matrix = ratings_table.pivot(index='movie_id',columns='user_id',values='rating').fillna(0)
```

```python
ratings_matrix.loc[1]
```

```python
ratings_matrix.head()
```

```python
movie_id = 1
ratings_for_movie_id = ratings_matrix.ix[movie_id]

# Filter 0 ratings 
ratings_for_movie_id = ratings_for_movie_id[ratings_for_movie_id > 0.0]
```

```python
users_who_have_seen_movie = ratings_for_movie_id.index
ratings_others = ratings_matrix[users_who_have_seen_movie]
```

```python
# Axis means - 1 or â€˜columnsâ€™ for row-wise
# Imagine each row in ratings_other being plotted against ratings_for_movie_id
movies_correlation = ratings_others.corrwith(ratings_for_movie_id, axis=1).dropna().sort_values(ascending=False)
```
